# Unilateral cross-feeding
[S1 Table - Unilateral parameters](https://doi.org/10.1371/journal.pcbi.1008135.s012)

## Model parameters

```julia
"""Parameters"""
unilateral_params = (V1g = 5.52e-13,
                    V3g = 6.28e-13,
                    Kg = 1e-12,
                    I1a = 2.63e1,
                    I3a = 1.13e2,
                    C1g = 8.7e-1,
                    V1a = 4.26e-13,
                    K1a = 2e-1,
                    γa = 2e11,
                    ψa = 3.3e-1,
                    δa = 3)
```

## Define model

```julia
#=
Calculation of uptake and growth rates
=#
function unilateralRates(A, G, p)
    # Uptake rates
    J1gu = (p.V1g * G)/(p.Kg + G)
    J3gu = (p.V3g * G)/(p.Kg + G)
    J1au = (p.V1a * p.C1g)/(p.K1a + A * p.C1g + G)
    
    # Growth rates
    J1grow = p.γa * ((1 - p.ψa) * p.δa * J1gu + J1au) * p.I1a / (p.I1a + A)
    J3grow = p.γa * (1 - p.ψa) * p.δa * J3gu * p.I3a / (p.I3a + A)
    
    return J1gu, J3gu, J1au, J1grow, J3grow
end

#= 
Unilateral cross-feeding between glucose and acetate specialists
Based on S1 Text 2.2
y = [G, N1, N3, A]
f = dy/dt
=#
function unilateral!(f, y, p, t)
    # Unpack y
    G, N1, N3, A = y
    
    # Uptake and growth rates
    J1gu, J3gu, J1au, J1grow, J3grow = unilateralRates(A, G, p)
    
    # dG/dt
    f[1] = p.D*(p.G0 - G) - J1gu * N1 - J3gu * N3
    
    # dN1/dt
    f[2] = N1 * (J1grow - p.D)
    
    # dN3/dt
    f[3] = N3 * (J3grow - p.D)
    
    # dA/dt
    f[4] = D * (p.A0 - A) + (p.ψa * p.δa * J1gu - J1au) * N1 + p.ψa * p.δa * J3gu * N3
    
    nothing
end
```

## Recreate 2F 
Growth rate ratio of CV101 to CV103 in the nutritional space. The gray shading indicates when CV101 grows faster than CV103 and the gray circles mark when their growth rates are both equal to the dilution rate 0.2 h-1

![](images/fig2.png)

- [Script](https://github.com/liaochen1988/Source_code_for_cross_feeding_paper/blob/master/unilateral_cross_feeding_model/gen_Fig2F.mhttps://github.com/liaochen1988/Source_code_for_cross_feeding_paper/blob/master/unilateral_cross_feeding_model/gen_Fig2F.m)
- [Model](https://github.com/liaochen1988/Source_code_for_cross_feeding_paper/blob/master/unilateral_cross_feeding_model/acetate_mediated_cf_simplified_model.m) (How is `t` used???)

```julia
"""Hyperparameters"""
# Dilution
D = .2
# Initial acetate and glucose concentrations
A0s = G0s = 10 .^ (-3:.01:3)
# Time
tf = 1

"""Parameter space"""
# Initialize matrix for saving growth rate ratios
ratios = zeros(length(G0s), length(A0s))

for i = 1:length(G0s)
    G0 = G0s[i]
    for j = 1:length(A0s)
        A0 = A0s[j]
        # Add params
        hyperparams = (D = D, A0 = 0, G0 = 0)
        p = merge(unilateral_params, hyperparams)
        # Initial conditions
        # G, N1, N3, A
        
        # Calculate minimum growth rate ratio
        ## Assuming they use the final growth rates
        growth_rates = unilateralRates(A0, G0, p)
        # J1grow / J3grow
        ratio = log(10, growth_rates[4] / growth_rates[5])
        ratios[i, j] = ratio
    end
end
```

```julia
#= Visualize =#
contour(G0s, A0s, ratios[end:-1:1, :], fill=true, levels=25, c=:bwr, clims=(-.5, .5))
xaxis!(:log)
yaxis!(:log)
xlabel!("log Acetate (mM)")
ylabel!("log Glucose (mM)")
```
